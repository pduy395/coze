```python
from selenium import webdriver
from selenium.webdriver.common.by import By

options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
dr = webdriver.Chrome(options=options)

dr.get("https://python.langchain.com/v0.1/docs/use_cases/web_scraping/") # Website used for scraping

#Displaying the title of the website in this case I had used GFG's Website
print(dr.title,"\n")

#Displaying some GFG's Articles
total_text = ""
for i in dr.find_elements(By.TAG_NAME,'p'):
  text = i.text
  total_text += text
  total_text += "\n"
# print(total_text)
dr.quit()


html_content = dr.page_source
print(html_content)

#quitting the browser
```
